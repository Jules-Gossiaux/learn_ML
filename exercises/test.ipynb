```python
import numpy as np
import matplotlib.pyplot as plt
import random

random.seed(42)  # Pour reproductibilité

BUDGET_INITIAL = 100_000
N_PARTIES = 100
SOUS_BUDGET = 1_000
SOUS_BUDGET_MAX = 1_100
PROBA_GAGNER = 0.49

def simulate_sub_game(sous_budget=SOUS_BUDGET, sous_budget_max=SOUS_BUDGET_MAX, proba=PROBA_GAGNER):
    mise = 1
    stack = sous_budget
    while True:
        if mise > stack:
            # On ne peut plus miser, partie perdue
            return stack - sous_budget  # Perte (toujours négative ou nulle)
        stack -= mise
        if random.random() < proba:
            # Gagné
            stack += mise * 2
            if stack >= sous_budget_max:
                return stack - sous_budget  # Gain (toujours >= 100)
            mise = 1
        else:
            # Perdu
            mise *= 2

def main():
    budget = BUDGET_INITIAL
    gains = []
    parties_gagnantes = 0
    parties_perdantes = 0
    evolution = [budget]

    for i in range(N_PARTIES):
        gain = simulate_sub_game()
        gains.append(gain)
        budget += gain
        evolution.append(budget)
        if gain >= 100:
            parties_gagnantes += 1
        else:
            parties_perdantes += 1

    print(f"Budget initial : {BUDGET_INITIAL}")
    print(f"Budget final   : {budget}")
    print(f"Gain total     : {budget - BUDGET_INITIAL}")
    print(f"Parties gagnantes : {parties_gagnantes}")
    print(f"Parties perdantes : {parties_perdantes}")

    plt.figure(figsize=(12, 6))
    plt.plot(evolution, marker='o')
    plt.title("Évolution du budget au fil des parties")
    plt.xlabel("Numéro de la partie")
    plt.ylabel("Budget")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
