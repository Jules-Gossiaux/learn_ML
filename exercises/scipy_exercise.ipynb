```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy import fftpack
```

```python
# En réalité x=t
t = np.linspace(0, 10, 500)
y =  2 * np.sin(2 * np.pi * 0.5 * t) + np.random.randn(t.size) * 0.5
```

```python
#Affichage de ce graphique bruité
plt.figure(figsize=(12, 8))
plt.title("Graph comprenant les bruits")
plt.plot(t, y)
plt.xlabel("axe x (t)")
plt.ylabel("axe y ")
plt.show()
```

```python
#Supprimer la tendance linéaire du signal

new_y = signal.detrend(y)
plt.figure(figsize=(12, 8))
plt.title("Grapgh sans les bruits")
plt.xlabel("axe x")
plt.ylabel("axe y")
plt.plot(t, new_y)
plt.show()

#Affichage de deux graph sans les bruits et avec le spectre


plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.title("Spetre du signal sans les bruits")
#Suppression des bruits
fourier = fftpack.fft(y)
power = np.abs(fourier)
frequences = fftpack.fftfreq(y.size)
fourier[power<100] = 0
plt.plot(np.abs(frequences), np.abs(fourier))

#Affichage du signal initial sans les bruits
plt.subplot(1, 2, 2)
plt.title("Affichage du signal sans les bruits")
filtered_signal = fftpack.ifft(fourier)
plt.plot(t, y, alpha=0.5, color="red")
plt.plot(t, filtered_signal, lw=3, color="green")
plt.show()
```
