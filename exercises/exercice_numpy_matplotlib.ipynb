```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Etape 1

```python
#Simulation de donn√©es
days = 100
TV_table = np.random.randint(10, 300, days)
radio_table = np.random.randint(0, 100, days)
newspaper_table = np.random.randint(0, 100, days)

#Generation of a sales variable(sales = 0.05*TV + 0.1*radio + 0.01*journaux + bruit)(Le bruit est un peu de np.random.normal(0, 2, 100))
noise = np.random.normal(0, 2, 100)
sales = 0.05*TV_table + 0.1*radio_table + 0.01*newspaper_table + noise
```

#### Etape 2

```python
# Moyenne des donn√©es et √©cart type
# Calculation of average
TV_mean = TV_table.mean()
radio_mean = radio_table.mean()
newspaper_mean = newspaper_table.mean()
sales_mean = sales.mean()
#Calculation of stadard deviation
TV_std = TV_table.std()
radio_std = radio_table.std()
newspaper_std = newspaper_table.std()
sales_std = sales.std()

print(f"moyenne et √©cart type de la TV: {TV_mean}, {TV_std}")
print(f"moyenne et √©cart type de la radio: {radio_mean}, {radio_std}")
print(f"moyenne et √©cart type de la newspaper: {newspaper_mean}, {newspaper_std}")
print(f"moyenne et √©cart type de la sales: {sales_mean}, {sales_std}")

#Calculation of correlations
TV_cor = np.corrcoef(TV_table, sales)
radio_cor = np.corrcoef(radio_table, sales)
newspaper_cor = np.corrcoef(newspaper_table, sales)
print(f"les corr√©lations entre la TV, radio, journal et les ventes sont: {TV_cor}, {radio_cor}, {newspaper_cor}")
```

#### Etape 4

```python
# Display of the datas

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.title("Histogram of TV")
plt.hist(TV_table, alpha=0.5, label="TV")
plt.subplot(2, 2, 2)
plt.title("Histogram of TV")
plt.hist(radio_table, alpha=0.5, label="radio")
plt.subplot(2, 2, 3)
plt.title("Histogram of TV")
plt.hist(newspaper_table, alpha=0.5, label="newspaper")
plt.show()

plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.title("correlation between TV ads and sales")
plt.scatter(TV_table, sales)
plt.xlabel("TV's budget")
plt.ylabel("Sales")
plt.subplot(2, 2, 2)
plt.title("correlation between radio ads and sales")
plt.scatter(radio_table, sales)
plt.xlabel("radio's budget")
plt.ylabel("Sales")
plt.subplot(2, 2, 3)
plt.title("correlation between newspaper ads and sales")
plt.scatter(newspaper_table, sales)
plt.xlabel("newspaper's budget")
plt.ylabel("Sales")
plt.show()


plt.figure(figsize=(12, 8))
plt.title("Sales curbes")
plt.plot(sales)
plt.show()
```
