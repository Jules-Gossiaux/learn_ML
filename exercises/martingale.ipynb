```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def game(stack=1000):
    bet = 1
    evolution_stack = []
    while stack > 0:
        stack -= bet
        if random.randint(0, 100) < 49:
            stack += bet * 2
            bet = 1
        else:
            bet = bet * 2
        if stack <= 0:
            break
        evolution_stack.append(stack)
    return np.array(evolution_stack)

def graph(data):
    evolution = data
    max_value = np.max(evolution)

    plt.figure(figsize=(12, 8))
    plt.title("Evolution du stack")
    plt.xlabel("Nombre de jeux")
    plt.ylabel("Stack")
    # Optionnel: marquer le point maximum sur le graphique
    plt.plot(np.argmax(evolution), max_value, 'ro', label='Maximum') # 'ro' = point rouge
    plt.plot(evolution)
    plt.show()

    print(evolution, max_value)
```

```python
def loop(number):
    table_max_values = np.array([])
    under_1100 = []
    for i in range (number):
        evolution = game()
        table_max_values = np.append(table_max_values, np.max(evolution))
        if np.max(evolution)>1100:
            under_1100.append(evolution)
        print(f"{i+1}. {table_max_values[i]}")
    print(len(under_1100))
    return table_max_values
data = loop(100)
plt.figure(figsize=(100, 100))
plt.hist(data, bins=10, color="red")

plt.title("Distribution des values max")
plt.xlabel("Pic du stack")
plt.ylabel("Nombre d'occurrences")
xmin, xmax = plt.xlim()
plt.xticks(np.arange(int(xmin), int(xmax)+5000, 5000))

plt.show()
```

```python
import random

def jeu(stack=1000, seuil=2000):
    mise = 1
    while 0 < stack < seuil:
        stack -= mise
        if random.randint(0, 100) > 51:  # 45% de chance de gagner
            stack += mise * 2
            mise = 1
        else:
            mise *= 2
        if mise > stack:
            break
    return stack

def compter_gains(n=100):
    compteur = 0
    for _ in range(n):
        resultat = jeu()
        if resultat >= 1100:
            compteur += 1
    return compteur

# Exemple d'utilisation :
nb_parties = 1000
victoires = compter_gains(nb_parties)
print(f"Nombre de fois que le budget a dpass 1100 sur {nb_parties} parties :", victoires)
```

```python
import numpy as np
import random

def jeu(stack=1000, mise_max=3000, objectif=1100):
    mise = 1
    while stack > 0 and stack < objectif:
        if mise > mise_max:
            mise = 1
        if stack < mise:
            break
        stack -= mise
        if random.random() < 51:  # 45% chance de gagner
            stack += mise * 2
            mise = 1
        else:
            mise *= 2
    return stack > objectif

def simulation(n=100):
    gains = sum(jeu() for _ in range(n))
    print(f"Sous-parties gagnes (stack > 1100) : {gains}/{n}")
    return gains

# Exemple :
simulation()
```
