### Cycle de vie d'une figure
<img src="images/cycles_de_vie_matplotlib.png" alt="Explication du code" width="300">

```python
# Convntionnellement on crée une figure de cette façon:
#Création de la "feuille de travail" => en param: la taille de la figure
plt.figure(figsize=(12, 8))
#création du graphique
plt.plot(x, y, label="quadratique")
#On peut rajouter d'autres figures sur la même feuille de travail
plt.plot(x, x**3, label="cubique")
#Ajout d'un titre
plt.title("figure 1")
#Ajout d'un nom sur l'axe des abscisses
plt.xlabel("axe x")
#Ajout d'un nom sur l'axe des ordonnées
plt.ylabel("axe y")
#Ajout d'une légende
plt.legend()
#Terminer le graphique (bien que ce ne soit pas obligatoire dans jupyter)
plt.show()
#Sauvegarder la figure
plt.savefig("figure.png")

#On peut aussi créer plusieurs figures sur la même feuille de travail
plt.subplot(2, 1, 1) #2 lignes, 1 colonne, 1er graphique
plt.plot(x, y, c="red", label="quadratique")
plt.subplot(2, 1, 2) #2 lignes, 1 colonne, 2ème graphique
plt.plot(x, np.sin(x), label="sinus")
plt.plot(x, np.cos(x), label="cosinus")
plt.legend()
plt.show()
```

#### Exercice perso: 
Créer un graph qui montre l'évolution d'un portefeuille sur un jeu aléatoire à deux issues équiprobables(pile ou face)

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def game(stack=1000):
    bet = 1
    evolution_stack = np.array([])
    while stack>0:
        stack -= bet
        if random.randint(0, 1) == 1:
            stack += bet*2
            bet = 1
        else:
            bet = bet*2
        if stack <= 0:
            break
        evolution_stack = np.append(evolution_stack, stack)
    return evolution_stack
game()
plt.figure(figsize=(12, 8))
plt.title("Evolution du stack")
plt.xlabel("Nombre de jeux")
plt.ylabel("Stack")
plt.plot(game())
```

#### Top5: plt_scatter

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

x = iris.data
y = iris.target
names = list(iris.target_names)

print(f'x contient {x.shape[0]} exemples et {x.shape[1]} variables')
print(f'il y a {np.unique(y).size} classes')

plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5, s=x[:, 2]*100)
```

### Top4: graphique en 3D

```python
from mpl_toolkits.mplot3d import Axes3D
ax = plt.axes(projection='3d')
ax.scatter(x[:, 0], x[:, 1], x[:, 2], c=y)
```

```python
f = lambda x, y: np.sin(x) + np.cos(x+y)
X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)

X, Y = np.meshgrid(X, Y)
Z = f(X, Y)
print(Z.shape)
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap="plasma")
```
