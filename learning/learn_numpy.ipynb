## Méthode permettant de manipuler un tableau

```python
I = np.zeros((3, 2))
J = np.ones((3, 2))

print(I)
print(J)
```

```python
#Assembler les tableaux soit de façon verticale soit horizontale
#De façon horizontale:
new_table1 = np.hstack((I, J))
print(new_table1)
print(new_table1.shape, new_table1.size)

#De façon horizontale:
new_table2 = np.vstack((I, J))
print(new_table2)
print(new_table2.shape, new_table2.size)

#La manière est choisie selon le param(1 pour horizontale et 0 pour verticale)
new_table3 = np.concatenate((I, J), axis=0)
print(new_table3, new_table3.shape, new_table3.size)
```

```python
#Changements de shape d'un tableau

J = J.reshape((1, 6))

K = np.array([1, 2, 3])
#Cette modification modifie la shape car dans certains cas cela pose problème que lorsqu'il n'y a qu'une ligne il n'y ai qu'une info dans le tuple. 
K = K.reshape(1, (K.shape[0]))
K = np.concatenate((K, np.zeros((1, 3))), axis = 0)
print(K, K.size, K.shape)

L = np.array([1, 2, 3])
print(L, L.shape)
L = L.reshape(1, (L.shape[0]))
print(L.shape)

#Retourne une shape avec un seul nombre lorsque le tableau ne possède qu'une seule ligne
L = L.squeeze()
print(L.shape)

#Retoiurne un tableau à une seule dimension
M = np.zeros((3, 4))
print(M)
print(M.ravel())
```

### Nan correction

```python
#Création d'un tableau d'exercice
import random
U = np.random.randn(5, 5)
random.seed(3)
for i in range(2):
    U[random.randint(0, 4), random.randint(0, 4)] = "nan"
print(U)

#Calculer la moyenne du tableau en ignorant les nan
np.nanmean(U)
#Calcule l'écart type en ignorant les nan
np.nanstd(U)
###Compte le nombre de nan dans le tableau
#retourne un tableau de la même forme que U avec des True et False selon si il y a un nan ou pas
np.isnan(U)
#Fait la somme de tous les True(et renvoie)(et ensuite calcule le rapport en %)
np.isnan(U).sum() / U.size

#Change les nan en 0
U[np.isnan(U)] = 0
print(U)
```

##### Exercices statistiques 

```python
np.random.seed(0)
V = np.random.randint(0, 100, (10, 5)).astype(np.float64)
print(V)


for a in range(V.shape[1]):
    means = [V[:, i].mean() for i in range(V[a].size)]
    stds = [V[:, i].std() for i in range(V[a].size)]
    new_data = (V[:, a]-means[a])/stds[a]
    V[:, a] = new_data

#Permet de supprimer les exposants et de garder seulement deux chiffres après la virgule
np.set_printoptions(precision=2, suppress=True)
print(V)

######Correction:
D = (V - V.mean(axis=0))/V.std(axis=0)
print(D)
```

## Broadcasting

Permet d'additionner deux tableaux en écrivant simplement:
C = A+B

```python
np.random.seed(1)
W = np.random.randint(0, 10, (2, 3))
X = np.ones((2, 3))
#additionne les deux tableaux
W+X
print(W)
Y = np.full((2, 1), 2)
#Ajoute sur chaque colonne le tableau Y
print(W+Y)
#S'il on a un tableau (1, 4) et (3, 1) on ontient unn tableau (3, 4)
```

#### Le broadcasting consiste à étendre les dimensions d'un tableau afin de pouvoir additionner deux tableaux de différentes dimensions
