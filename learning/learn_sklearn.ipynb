### 1) Régression avec SKLearn

```python
#importe le modèle de régression linéaire
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
m = 100 # number of samples
#Précision de 2 dimentions
X = np.linspace(0, 10, m).reshape(m, 1)
y = X + np.random.randn(m, 1)
```

```python
plt.scatter(X, y)
```

```python
#recopiage des 4 lignes précédemment vues(pas besoin d' hyperparamètre car utilise moindres carrés qui n'utilisents pas de paramètres)
model = LinearRegression()
model.fit(X, y)
#Affiche le score du modèle(coefficient de détermination R^2)
model.score(X, y)
predictions = model.predict(X)

plt.scatter(X, y)
plt.plot(X, predictions, color="red")
```
