### 1) Rgression avec SKLearn

```python
#importe le modle de rgression linaire
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
m = 100 # number of samples
#Prcision de 2 dimentions
X = np.linspace(0, 10, m).reshape(m, 1)
y = X + np.random.randn(m, 1)
```

```python
plt.scatter(X, y)
```

```python
#recopiage des 4 lignes prcdemment vues(pas besoin d' hyperparamtre car utilise moindres carrs qui n'utilisents pas de paramtres)
model = LinearRegression()
model.fit(X, y)
#Affiche le score du modle(coefficient de dtermination R^2)
model.score(X, y)
predictions = model.predict(X)

plt.scatter(X, y)
plt.plot(X, predictions, color="red")
```
